```python
import pandas as pan 
import numpy as num
import matplotlib.pyplot as plt
import seaborn 
data_frame=pan.read_csv('Weather Dataset.csv')
data_frame.drop_duplicates(inplace=True)
data_frame.dropna(inplace=True)
data_frame=data_frame[data_frame['Temperature (C)']>-75]
data_frame=data_frame[data_frame['Apparent Temperature (C)']>-75]
data_frame=data_frame[data_frame['Humidity']>0]
data_frame=data_frame[data_frame['Wind Speed (km/h)']>10]
data_frame=data_frame[data_frame['Wind Bearing (degrees)']>10]
data_frame= data_frame.sort_values(by=['Formatted Date'],ascending=True) 
x_axis = data_frame['Formatted Date']
y_axis = data_frame['Temperature (C)']
scater1=data_frame['Humidity']
plt.plot(x_axis, y_axis)
plt.show()
plt.hist(y_axis)
plt.show()
plt.scatter(x_axis,scater1)
plt.show()
data_frame = data_frame.select_dtypes(include=['float64', 'int64'])
seaborn.heatmap(data_frame.corr())
plt.show()



```
